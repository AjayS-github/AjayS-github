# Maxwellian velocity distribution
If a gas is in thermodynamic equilibrium with temperature $T$, the number of particles having speeds between $v$ and $v+dv$ is 

 $$ dn_v=4\pi n\left(\dfrac{m}{2\pi kT}\right)^{3/2} v^2 \exp\left(\dfrac{-mv^2}{2kT}\right) dv, $$
where $n$ is number of particles per unit volume.<br/>
$m$ is mass of each particle.<br/>
$k$ is Boltzmann constant.<br/>
$T$ is Temperature.

Let us plot the above expression for three different temperatures $T_1, T_2 \ \text{and}\  T_3$.
Let us say the gas is hydrogen in which case, $m=1.67\times10^{-27}$kg. <br/>
Python code for plotting the graph is shown below along with the output. <br/>
Notice that these are Gaussian curves. Observe how the maximum of the curve shifts along horizontal axis and height of the curve falls with increase in temperature.

```python
import numpy as np
import matplotlib.pyplot as plt

T1=300
T2=400
T3=500
m=1.67e-27
k=1.38e-23
n=6.23e23
v=np.arange(0,1e4,100)
n1=(4*np.pi*n*(m/(2*np.pi*k*T1))**(3/2))*v**2*np.exp(-m*v**2/(2*k*T1))
n2=(4*np.pi*n*(m/(2*np.pi*k*T2))**(3/2))*v**2*np.exp(-m*v**2/(2*k*T2))
n3=(4*np.pi*n*(m/(2*np.pi*k*T3))**(3/2))*v**2*np.exp(-m*v**2/(2*k*T3))
plt.figure(figsize=(10,5))
plt.plot(v,n1,'1',lw=1,label="T1=300K")
plt.plot(v,n2,'.',lw=1,label="T2=400K")
plt.plot(v,n3,'*',lw=1,label="T3=500K")
plt.ylim(0,0.3e21)
plt.xlim(-1e-1,1e4)
plt.xlabel('velocity')
plt.ylabel('Number of particles')
plt.title("Maxwellian velocity distribution")
plt.legend()
plt.show()
```
